```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')

frame_len = 1000
fig =plt.figure(figsize=(8,5))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['india']
    y2 = data['china']
    y3 = data['pakistan']
    if len(y1) <= frame_len:
        plt.cla()
        plt.plot(y1,label='India')
        plt.plot(y2,label='China')
        plt.plot(y3,label='Pakistan')
    else:
        plt.cla()
        plt.plot(y1[-frame_len:],label='India')
        plt.plot(y2[-frame_len:],label='China')  
        plt.plot(y3[-frame_len:],label='Pakistan')
    
    plt.title("India vs China vs Pakistan Twitter Sentiment Analysis")
    plt.legend(loc='upper-left')
    plt.tight_layout()
ani = FuncAnimation(plt.gcf(),animate,interval=1000)
```
